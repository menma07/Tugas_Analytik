```python
import pandas as pd

covid_df = pd.read_csv('COVID-19 Coronavirus.csv')
    
print(covid_df.head())
import matplotlib.pyplot as plt
```

#Scatter Plot

```python
plt.scatter(covid_df["Population"], covid_df["Death percentage"])
plt.title("Country Population and COVID Death Percentage")
plt.xlabel("Population")
plt.ylabel("Death Percent")
plt.show()
```

#Line Plot

```python
plt.plot(covid_df["Population"], covid_df["Death percentage"])
plt.title("Country Population and COVID Death Percentage")
plt.ylabel("Population")
plt.xlabel("Death Percent")
plt.show()
```

#Histogram

```python
plt.hist(covid_df["Death percentage"])
plt.title("Percentages of COVID Deaths by Country")
plt.show()
```

#Box and Whisker

```python
plt.boxplot(covid_df["Death percentage"])
plt.title("COVID Death Percentages")
plt.show()
```

#Pie Chart

```python
plt.pie(covid_df["Death percentage"])
plt.title("COVID Death Percentage")
plt.show()
```
